```python
imagelist = os.listdir('/content/Image') #lists the images in the directory
```

```python
segments=[]
for i in range(len(imagelist)):
  img = imread('/content/Image/'+imagelist[i])
  img = sq.im.ImageContainer(img)
  sq.im.segment(
    img=img,
    layer="image",
    channel=0,
    method=stardist_2D_versatile_fluo,
    layer_added="segmented_stardist",
    nms_thresh=None,
    prob_thresh=None,
  )
  print(img)
  segments.append(img)
```

```python
for i in range(len(segments)):
  img = segments[i]
  fig, axes = plt.subplots(1, 2)
  img.show("image", channel=0, ax=axes[0])
  _ = axes[0].set_title("DAPI")
  img.show("segmented_stardist", cmap="jet", interpolation="none", ax=axes[1])
  _ = axes[1].set_title(f"Number of segments: {len(np.unique(img['segmented_stardist']))}")
```
